```python
import pandas as pd
import sklearn as sk
```

```python
df = pd.read_csv("spam-datasset.csv")
df
```

```python
df['Category'] = df['Category'].replace({'spam':1, 'ham':0})
df
```

## Separando em train e test

```python
from sklearn.model_selection import train_test_split

x_train, x_test,y_train, y_test = train_test_split(df['Message'],df['Category'],test_size = 0.2)
x_train
```

## Modelo Bag of Words

```python
# Esse processo de converter um vetor textual em um vetor númerico é chamado de vectorizer
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
features = vectorizer.fit_transform(x_train)
features
```

```python
features.toarray()
```

```python
vectorizer.get_feature_names()
```

## Construindo o Modelo

```python
from sklearn import svm 

## Vamos usar o SVM para este modelo
modelo = svm.SVC()
modelo.fit(features,y_train)
```

```python
testeFeatures = vectorizer.transform(x_test)
print("Accuracy: {}".format(modelo.score(testeFeatures,y_test)))
```

```python
## Criando uma função para facilitar nosso trabalho futuro

def criarModelo(vectorizer):
    features = vectorizer.fit_transform(x_train)
    modelo = svm.SVC()
    modelo.fit(features,y_train)
    testeFeatures = vectorizer.transform(x_test)
    print("Accuracy: {}".format(modelo.score(testeFeatures,y_test)))

criarModelo(CountVectorizer())
```

## Usando o TFidVectorizer


```python
from sklearn.feature_extraction.text import TfidfVectorizer

TFIDvectorizer = TfidfVectorizer()

X = TFIDvectorizer.fit_transform(x_train)
print("Shape da matriz: ",X.shape)
TFIDvectorizer.get_feature_names()

```

```python
criarModelo(TfidfVectorizer())
```

## Testando com Hashing Vectorizer

```python
from sklearn.feature_extraction.text import HashingVectorizer

criarModelo(HashingVectorizer(n_features=2**8))
```

```python
# Podemos ver que a medida que o número de hashs aumenta, a acurácia segue aumentando. até um ponto onde ela se estabiliza
for i in range(15):
    criarModelo(HashingVectorizer(n_features=2**i))
```
